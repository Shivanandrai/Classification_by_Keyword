# 1 | Dependent Variable "Sex"

1.1 Comparison of Algorithms

```
dependent_variable_sex = {'M': 1,'F': 0}
data.dependent_variable_sex = [dependent_variable_sex[item] for item in data.dependent_variable_sex]
```

```
X_sex = data.drop(['dependent_variable_sex','dependent_variable_age','dependent_category_age'], axis=1)
Y_sex= data['dependent_variable_sex']
X_train_sex, X_validation_sex, Y_train_sex, Y_validation_sex = train_test_split(X_sex, Y_sex, test_size=0.20, random_state=123, stratify= Y_sex)

X_train_sex.set_index(['ID'])
X_validation_sex.set_index(['ID'])
```

```
estimator =  100
models=[]
models.append(('GBC',GradientBoostingClassifier(n_estimators = estimator, random_state = 123)))
models.append(('RFC',RandomForestClassifier(n_estimators= estimator)))
models.append(('CART',DecisionTreeClassifier()))
models.append(('KNN',KNeighborsClassifier()))
models.append(('NNB',GaussianNB()))
models.append(('LR',LogisticRegression()))


# evaluate each model in turn
results = []
names = []
for name, model in models:
    model = model.fit(X_train_sex, Y_train_sex)
    trainpred = model.predict(X_train_sex)
    print(name)
    print('Accuracy score')
    print(metrics.accuracy_score(Y_train_sex,trainpred))
    print('Report')
    print(metrics.classification_report(Y_train_sex,trainpred))
    
#Evaluate predictions
    y_pred =model.predict(X_validation_sex)
    print('Prediction Accuracy score')
    print(metrics.accuracy_score(Y_validation_sex,y_pred))
    print('Confusion Matirx')
    print(pd.crosstab(y_pred, Y_validation_sex, 
            rownames=['pred'], 
            colnames=['Actual'], margins=False, margins_name="Total"))
    print('Prediction Report')
    print(metrics.classification_report(Y_validation_sex,y_pred))
```

1.2 Optimization of Algorithm

```
rf = RandomForestClassifier()

pprint(rf.get_params())
```

```
# Number of trees in random forest
n_estimators = [50, 100, 150]
# Number of features to consider at every split
max_features = ['auto']
# Maximum number of levels in tree
max_depth = [5, 10, 20, 30, 40]
# Minimum number of samples required to split a node
min_samples_split = [200, 250, 300]
# Minimum number of samples required at each leaf node
min_samples_leaf = [5, 10, 15, 20, 25]
# Method of selecting samples for training each tree
bootstrap = [True]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
pprint(random_grid)
```

```
rf = RandomForestClassifier()
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=123, n_jobs = -1)
rf_random.fit(X_train_sex, Y_train_sex)

rf_random.best_params_
```

```
base_model = RandomForestClassifier(n_estimators = 100, random_state = 123)
base_model.fit(X_train_sex, Y_train_sex)
base_predictions= base_model.predict(X_validation_sex)
best_random = rf_random.best_estimator_
Best_predictions= best_random.predict(X_validation_sex)


improvement= metrics.accuracy_score(Y_validation_sex, Best_predictions)- metrics.accuracy_score(Y_validation_sex,base_predictions)

print(improvement)
print(metrics.accuracy_score(Y_validation_sex, Best_predictions))
```

2.1.1 Regression-Algorithms - Comparison of Algorithms

```
X_train, X_test, y_train, y_test = train_test_split(data[cols[4:]], np.array(data[cols[1:2]]).flatten(), test_size=0.25)
```

```
skf = StratifiedKFold(n_splits=5,shuffle=False)
GBR=GradientBoostingRegressor(loss='ls', learning_rate=0.1, n_estimators=50, criterion='friedman_mse', min_samples_split=2, min_samples_leaf=10)
RFR=RandomForestRegressor(n_estimators=50,criterion="mse",min_samples_leaf=10)
LR=LinearRegression()
BR=BayesianRidge()
```

```
scores_GBR=cross_validate(GBR,X=X_train,y=y_train,return_train_score=True,cv=skf,scoring=["r2","neg_mean_absolute_error"])
scores_GBR = pd.DataFrame(scores_GBR)
print("1/4 done")
scores_RFR=cross_validate(RFR,X=X_train,y=y_train,return_train_score=True,cv=skf,scoring=["r2","neg_mean_absolute_error"])
scores_RFR = pd.DataFrame(scores_RFR)
print("2/4 done")
scores_LR=cross_validate(LR,X=X_train,y=y_train,return_train_score=True,cv=skf,scoring=["r2","neg_mean_absolute_error"])
scores_LR = pd.DataFrame(scores_LR)
print("3/4 done")
scores_BR=cross_validate(BR,X=X_train,y=y_train,return_train_score=True,cv=skf,scoring=["r2","neg_mean_absolute_error"])
scores_BR = pd.DataFrame(scores_BR)
print("4/4 done")
```

```
regression_CV = np.concatenate((np.array(scores_GBR.mean()),
                               np.array(scores_RFR.mean()),
                               np.array(scores_LR.mean()),
                               np.array(scores_BR.mean())),
                               axis=0)
regression_scores = pd.DataFrame(regression_CV.reshape(4,6),
                                 columns=scores_GBR.columns,
                                 index=["scores_GBR","scores_RFR","scores_LR","scores_BR"])
regression_scores
#Random Forest seems to be the best
```
